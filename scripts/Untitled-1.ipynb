```python
import matplotlib.pyplot as plt
import json

# Provided JSON data
file  = "/projects/COMPXR/pranay/Eyes/compxr/results/00009-stylegan3-t-OCT_Train-gpus2-batch16-gamma2/stats.jsonl"

# Parse the JSON data
records = json.load(file)

# Extract FID scores and corresponding snapshot names
fid_scores = [record[' results']['fid50k_full'] for record in records]
snapshots = [record['snapshot_pkl'] for record in records]

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(snapshots, fid_scores, marker='o')
plt.xlabel('Snapshot')
plt.ylabel('FID Score')
plt.title('FID Score over Different Snapshots')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import os
import shutil

# Set these variables
csv_file_path = '/projects/COMPXR/pranay/Eyes/Datasets/Experiment/image_data.csv'  # Path to your CSV file
source_base_path = '/projects/COMPXR/pranay/Eyes/Datasets/OCT/train'  # Base path where images currently are
destination_base_path = '/projects/COMPXR/pranay/Eyes/Datasets/Experiment_2'  # Base path where you want to copy images

# Read the CSV file
df = pd.read_csv(csv_file_path)

# Loop through the DataFrame
for index, row in df.iterrows():
    # Extracting the file path from the CSV
    file_path = row['Original Path'] 
    new_filename = row['New Filename']

    # Generating the new source and destination paths
    class_name = os.path.basename(os.path.dirname(file_path))
    img_name = os.path.basename(file_path)
    new_source_path = os.path.join(source_base_path, class_name, img_name)
    new_destination_path = os.path.join(destination_base_path, new_filename)

    # Create directory if it doesn't exist
    os.makedirs(os.path.dirname(new_destination_path), exist_ok=True)

    # Copy the file
    if row['Type']== 'Original':
        shutil.copy2(new_source_path, new_destination_path)
    else:
        print(f"skipped {new_source_path}")
```
